```python
dataframe = df.dropna()
```

```python
dataframe.head(30)
```

```python
dataframe['budget'] = dataframe['budget'].astype('int64')
dataframe['gross'] = dataframe['gross'].astype('int64')
```

```python
dataframe
```

```python
for col in dataframe.columns:
    pct_missing = np.mean(dataframe[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
dataframe = df.sort_values(by=['gross'],inplace=False,ascending = False)
```

```python
dataframe['company'].drop_duplicates().sort_values(ascending=False)
```

```python
dataframe
```

```python
plt.scatter(x=dataframe['budget'],y=dataframe['gross'])
plt.title('Budget VS Gross earning')
plt.show
```

```python
dataframe.head()
```

```python
sns.regplot(x=dataframe['budget'],y=dataframe['gross'],line_kws={"color":"green"})
```

```python
corelation_matrix = dataframe.corr(method = 'pearson')
```

```python
corelation_matrix
```

```python
sns.heatmap(corelation_matrix,annot = True)
plt.title('corelation_matrix')
plt.show()
```

```python
df_numerized = dataframe


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
num_corelation_matrix = df_numerized.corr(method = 'pearson')
```

```python
sns.heatmap(num_corelation_matrix,annot = True)
plt.title('numerised corelation_matrix')
plt.show()
```

```python
corelation_mat =  df_numerized.corr(method = 'pearson')
corr_pairs = corelation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs>0.5]
```
